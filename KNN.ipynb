```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
dataset=pd.read_csv("Social_Network_Ads.csv")
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split as ts
x_train,x_test,y_train,y_test=ts(x,y,test_size=0.25,random_state=0)
```

```python
print(x_train)
```

```python
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler 
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn=KNeighborsClassifier(n_neighbors= 5,metric='minkowski',p=2)
kn.fit(x_train,y_train)
```

```python
y_pred=kn.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm= confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Ensure that sc is properly defined before using inverse_transform
x_set, y_set = sc.inverse_transform(x_train), y_train

# Create meshgrid
x1, x2 = np.meshgrid(
    np.arange(start=x_set[:, 0].min() - 2, stop=x_set[:, 0].max() + 2, step=0.5),
    np.arange(start=x_set[:, 1].min() - 2000, stop=x_set[:, 1].max() + 2000, step=500)
)

# Predict and reshape
predictions = kn.predict(sc.transform(np.c_[x1.ravel(), x2.ravel()])).reshape(x1.shape)

# Contour plot
plt.contourf(x1, x2, predictions, alpha=0.75, cmap=ListedColormap(('red', 'green')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())

# Scatter plot of training data points
colors = ListedColormap(('red', 'green'))
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1], color=colors(i), label=j)

# Labels and title
plt.title("K-Nearest Neighbors (Testing set)")
plt.xlabel("Age")
plt.ylabel("Estimated Salary")
plt.legend()
plt.show()
```
